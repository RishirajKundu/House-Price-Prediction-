### Using Lasso to find most significant features

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha = 0.005, random_state = 0)) 
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_features = X_train.columns[(feature_sel_model.get_support())]
```

```python
X_train_lasso = X_train[selected_features]
X_test_lasso = X_test[selected_features]
```

###  Hyper param optimization using Randomized search CV

### XGBoost Regressor


```python
reg = xgboost.XGBRegressor()

boosters = ["gbtree", "gblinear"]
base_score = [0.25, 0.5, 0.75, 1]
n_estimators = [100, 500, 900, 1100, 1500]
max_depth = [2, 3, 5, 10, 15]
learning_rate = [0.05, 0.1, 0.15, 0.20]
min_child_weight = [1, 2, 3, 4]

hyperparameter_grid = {
    
    "n_estimators" : n_estimators,
    "max_depth" : max_depth,
    "learning_rate" : learning_rate,
    "min_child_weight" : min_child_weight,
    "boosters" : boosters,
    "base_score" : base_score    
}


from sklearn.model_selection import RandomizedSearchCV
r_cv = RandomizedSearchCV(estimator=reg, param_distributions=hyperparameter_grid, cv = 5, n_iter=50,
                         scoring="neg_mean_absolute_error", n_jobs= 4,
                         verbose=5, return_train_score=True, random_state=1)

r_cv.fit(X_train_lasso, y_train)
```

```python
r_cv.best_estimator_
```

```python
reg = xgboost.XGBRegressor(base_score=1, booster='gbtree', boosters='gblinear',
             colsample_bylevel=1, colsample_bynode=1, colsample_bytree=1,
             gamma=0, importance_type='gain', learning_rate=0.05,
             max_delta_step=0, max_depth=2, min_child_weight=1, missing=None,
             n_estimators=900, n_jobs=1, nthread=None, objective='reg:linear',
             random_state=0, reg_alpha=0, reg_lambda=1, scale_pos_weight=1,
             seed=None, silent=None, subsample=1, verbosity=1)

reg.fit(X_train_lasso, y_train)
```

```python
pred = reg.predict(X_test_lasso)
```

```python
sub_df = pd.read_csv("sample_submission.csv")
df = pd.concat([sub_df["Id"], pd.DataFrame(pred, columns = ["SalePrice"])], axis = 1)
```

```python
df["SalePrice"] = np.exp(df["SalePrice"])
```
